```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
data = pd.read_csv('data1.txt', sep=',', header=None)
X = data[0]
Y = data[1]
m = Y.size
print (m)
```

```python
def plotData(x,y):
    fig = pyplot.figure()
    pyplot.plot(x,y,'ro', ms=10,mec='k')
    pyplot.ylabel('Y')
    pyplot.xlabel('X')
    pyplot.show()
```

```python
plotData(X,Y)
```

```python
X = np.stack([np.ones(m), X], axis=1)
```

```python
def costFunc(x,y,theta):
    m = Y.size
    J = 0
    J = (1/(2*m))*np.sum(np.square(np.dot(X,theta) - y))
    return J
```

```python
J = costFunc(X,Y, theta=np.array([0,0]))
print("with theta = [0,0] \nCost computed = %.2f" % J)
```

```python
J = costFunc(X,Y, theta=np.array([-2,4]))
print("with theta = [0,0] \nCost computed = %.2f" % J)
```

```python
def gradientDesc(x,y,theta, alpha, iters):
    m = y.size
    xt = x.transpose()
    for i in range(0, iters):
        h = np.dot(x, theta)
        loss = h - y
        cost = np.sum(loss**2)/(2*m)
        print("Iteraton %d | Cost: %f" % (i, cost))
        gradient = np.dot(xt, loss)/m
        theta = theta -alpha*gradient
    return theta
        
    
```

```python
theta = np.zeros(2) #initialize the values of theta
iterations = 1000 # iterations to find the optimal theta
alpha = 0.001 # learning rate
theta, J_history = gradientDesc(X ,Y, theta, alpha, iterations)
print('Theta found by gradient descent: {:.4f}, {:.4f}'.format(*theta))
plotData(X,Y)
```
